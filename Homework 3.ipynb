# Calculating Divident

```r
f<-function(x){
  r1=0
  r2=0
  d=0
  x$Dividend[1]='NA'
  for (i in 2:251){
    r1=x$Close[i-1]/x$Close[i]
    r2=x$Adj.Close[i-1]/x$Adj.Close[i]
    d=r1-r2
    x$Dividend[i]=x$Close[i]*d
}
  return(x)
}
```

# IBM

```r
ibm_data<-f(ibm_data)
ibm_data<-select(ibm_data,Date,Dividend)
head(ibm_data)
tail(ibm_data)
```

```r
ibm_data<-filter(ibm_data,Dividend!=0 & Dividend!='NA')
head(ibm_data)
str(ibm_data)
result<-filter(ibm_data,Date=='2018-02-08')
print("Divident value on  2/8/2018")
result
```
